```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes=pd.read_csv('diabetes.csv')
```

# Descripción de Datos

Pregnancies: Cuantas veces estuvo embarazada - cuantitativa, discreta   
Glucose: Nivel de azúcar en la sangre - cuantitativa, discreta   
BloodPressure: Presión Diastólica - cuantitativa, discreta   
SkinTickness: Delgadez de la piel - cuantitativa, discreta   
Insulin: Nivel de insulina en la sangre - cuantitativa, discreta    
BMI: Índice de Masa Corporal - cuantitativa, continua    
DiabetesPedigreeFunction:Probabilidad de diabetes según antecedences familiares - cuantitativa, continua   
Age:Edad de la persona - cuantitativa, discreta   
Outcome: Si tiene o no tiene diabetes - cualitativa, nominal (1=Si, 0=No)

# Visualización de Datos

```python
diabetes.head(10)
```

```python
fig = plt.figure(figsize=(10,5))
diabetes.Outcome.value_counts().plot(kind="bar",alpha=0.5)  
```

En la gráfica anterior podemos observar como más personas entrevestidas no presentan diabetes, ya que el 0 se refiere a las que no lo presentan.

```python
ax = sns.boxplot(x="Outcome", y="Age", 
                data=diabetes)
ax = sns.stripplot(x="Outcome", y="Age",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

Dentro de la gráfica anterior podemos observar como la mayoría de las edades se encuentra dentro del 50% de los datos, se tienen puntos atípicos que son los que están fuera de los bigotes. Además de que se puede observar que de los que no tienen diabetes la edad se concentra entre 20 y 40 y de los que si tienen se puede observar que están entre los 30 y 45. 

```python
ax = sns.boxplot(x="Outcome", y="Pregnancies", 
                data=diabetes)
ax = sns.stripplot(x="Outcome", y="Pregnancies",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

En la gráfica anterior podemos observar datos atípicos dentro de los que no tienen diabetes, ya que arriba de 11 embarazos se considera atípico, sin embargo en los que si tienen diabetes, podemos ver como 17 embarazos entran dentro del rango del diagrama. 

```python
ax = sns.boxplot(x="Outcome", y="Insulin", 
                data=diabetes)
ax = sns.stripplot(x="Outcome", y="Insulin",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

En la gráfica anterior podemos observar como existen muchos puntos atípicos en los valores de insulina, tanto en los que no tienen diabetes como en los que si, podemos ver mayores valores de insulina en los que si tienen ya que en la diabetes se necesita insulina para que la glucosa puede entrar a las celulas. 

```python
ax = sns.boxplot(x="Outcome", y="DiabetesPedigreeFunction", 
                data=diabetes)
ax = sns.stripplot(x="Outcome", y="DiabetesPedigreeFunction",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

En la gráfica anterior podemos observar como existen datos atípicos, más en los que no tienen diabetes, significa que tienen una descendencia familiar muy fuerte de diabetes, sin embargo aun no les pega, en los que si tienen diabetes, podemos ver como la descendencia familiar si ha afectado, ya que se tiene un rango mayor y los datos estan muy concentrados dentro del 50%.
