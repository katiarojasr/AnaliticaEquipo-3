# Patrones con KMeans

```python
diabetes_prep = diabetes.drop(['BloodPressure','SkinThickness'], axis=1)
```

```python
diabetes_prep = diabetes_prep.drop(['Outcome'], axis=1)
```

```python
# Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetes_prep.loc[:, ['Glucose', 'BMI']]
# KMEANS 1
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)

# visualization
plt.scatter(diabetes_prep['Glucose'], diabetes_prep['BMI'], c = labels)
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
# Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetes_prep.loc[:, ['Glucose', 'BMI']]
# KMEANS 2
from sklearn.cluster import KMeans
kmeans2 = KMeans(n_clusters=2)
kmeans2.fit(datos1)
labels2 = kmeans2.predict(datos1)
list_clusters = np.unique(labels2)
print("Número de clusters:",len(list_clusters))
print("Cluster Centers:",kmeans2.cluster_centers_)


# visualization
plt.scatter(diabetes_prep['Glucose'], diabetes_prep['BMI'], c = labels2)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.title('Glucose vs BMI')
plt.show()
```

```python
diabetes_nuevo=diabetes_prep.copy()

diabetes_nuevo.loc[:,'Outcome'] = diabetes['Outcome']
diabetes_nuevo.loc[:,'cluster'] = labels

diabetes_nuevo
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='Glucose',y='BMI',data=diabetes_nuevo, hue='Outcome',
           palette='coolwarm',height=6,aspect=1,fit_reg=False)
```
